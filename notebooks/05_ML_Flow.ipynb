```python
from pathlib import Path
import sys
import pickle

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Make sure we can import src.*
PROJECT_ROOT = Path("..").resolve()
if str(PROJECT_ROOT) not in sys.path:
    sys.path.insert(0, str(PROJECT_ROOT))

from src.feature_engineering import evaluate_model, log_model_to_mlflow

# Load test data
X_test = pd.read_csv("../data/processed/X_test.csv")
y_test = pd.read_csv("../data/processed/y_test.csv").values.ravel()

models_dir = PROJECT_ROOT / "models"
figures_dir = PROJECT_ROOT / "reports" / "figures"
figures_dir.mkdir(parents=True, exist_ok=True)
```

```python

with open(models_dir / "logistic_regression_tuned.pkl", "rb") as f:
    lr_t = pickle.load(f)
with open(models_dir / "decision_tree_tuned.pkl", "rb") as f:
    dt_t = pickle.load(f)
with open(models_dir / "random_forest_tuned.pkl", "rb") as f:
    rf_t = pickle.load(f)
with open(models_dir / "xgboost_tuned.pkl", "rb") as f:
    xgb_t = pickle.load(f)
```

```python
y_pred_lr_t = lr_t.predict(X_test);        y_proba_lr_t = lr_t.predict_proba(X_test)[:, 1]
y_pred_dt_t = dt_t.predict(X_test);        y_proba_dt_t = dt_t.predict_proba(X_test)[:, 1]
y_pred_rf_t = rf_t.predict(X_test);        y_proba_rf_t = rf_t.predict_proba(X_test)[:, 1]
y_pred_xgb_t = xgb_t.predict(X_test);      y_proba_xgb_t = xgb_t.predict_proba(X_test)[:, 1]
```

```python
# Log all models to MLflow
log_model_to_mlflow(lr_t, "Logistic_Regression", X_test, y_test, y_pred_lr_t, y_proba_lr_t)
log_model_to_mlflow(dt_t, "Decision_Tree", X_test, y_test, y_pred_dt_t, y_proba_dt_t)
log_model_to_mlflow(rf_t, "Random_Forest", X_test, y_test, y_pred_rf_t, y_proba_rf_t)
log_model_to_mlflow(xgb_t, "XGBoost", X_test, y_test, y_pred_xgb_t, y_proba_xgb_t)

print("Models successfully logged to MLflow! Run `mlflow ui` to view the results.")
```
